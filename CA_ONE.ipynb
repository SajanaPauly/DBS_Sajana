## **Package Importing**

Importing all the required packages for data analysis and web scraping.

```python
#Importing all the required packages
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from csv import writer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# defined four variables as empty lists

Product_name = []
Prices = []
Description = []
Reviews = []

# Constructs URL in each iteration and checking the response 
for i in range(2,12):

 url= "https://www.flipkart.com/search?q=mobiles%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off" + str(i)
 r = requests.get(url)
 print(r)

# BeautifulSoup to parse the HTML content and creating BeautifulSoup object 
 soup = BeautifulSoup(r.text,"lxml")
 box = soup.find("div", class_="_1YokD2 _3Mn1Gg")

# Searching div elements and iterates over the list.It then extract the content of each elememnt and append to the list 
 names = box.find_all("div",class_="_4rR01T")
 for i in names:
  name=i.text
  Product_name.append(name)
 print(Product_name)

 prices = box.find_all("div",class_="_30jeq3 _1_WHN1")
 for i in prices:
  name=i.text
  Prices.append(name)
 print(Prices)

 desc = box.find_all("ul",class_="_1xgFaf")
 for i in desc:
  name=i.text
  Description.append(name)
 print(Description)

 reviews = box.find_all("div",class_="_3LWZlK")
 for i in reviews:
  name=i.text
  Reviews.append(name)
 print(Reviews)
```

```python
columns = [Product_name, Prices, Description, Reviews]
max = 0
for i in columns:
  if len(i) > max:
    max = len(i)
print(max)

for j in columns:
  if len(j) < max:
    while len(j) != max:
      j.append('0')
```

```python
# CReating data frame 
a = {"Product Name":Product_name,"Prices":Prices,"Description":Description,"Reviews":Reviews}
df = pd.DataFrame(a)
df
```

This is the dataset without performing any transformations 

## **Data Cleaning**

Dataset cleaning is the process of fixing or removing errors, inconsistencies, and missing information from a dataset in order to prepare it for analysis. It is an important phase in the data preparation process because it can help to ensure that the analysis's findings are precise and meaningful.

```python

#Replace special charecters., in this case ₹
df["Prices"] = df["Prices"].str.replace('\W', '', regex=True)

#Split the columns into two with one having the details of memory and other having the features of the product
df[['Memory_Details', 'Features']] = df['Description'].str.split('|', 1, expand=True)

#Renaming the column Product Name to Product_Name
df.rename(columns={'Product Name': 'Product_Name'}, inplace=True)

#split the columns into two with containing product name and other having the color details
df[['Product_Name', 'Color']] = df['Product_Name'].str.split('(', 1, expand=True)

#creating a new column Brand which contains the name of the product brand
df['Brand'] = df['Product_Name'].str.split(' ').str[0]

#Adding the current date field
df['date'] = pd.to_datetime('today').strftime("%Y-%m-%d")
```
